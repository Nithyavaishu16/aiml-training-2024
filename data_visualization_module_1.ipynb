```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

```python
# Removing rows which have missing values
data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))
```

```python
# Dividing the dataset into features (X) and the target variable (y)

X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24, 7))
plt.scatter(X["make"], y)
plt.xlabel('make')
plt.ylabel('price')
plt.title('Car maker vs price - Scatter Plot')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "make": ["Toyota", "Honda", "Ford", "BMW", "Audi", "Toyota", "Honda", "Ford", "BMW", "Audi"],
    "body-style": ["sedan", "hatchback", "sedan", "convertible", "SUV", "SUV", "sedan", "SUV", "convertible", "hatchback"],
    "horsepower": [130, 150, 120, 300, 250, 200, 140, 210, 320, 160],
    "price": [22000, 18000, 25000, 35000, 50000, 45000, 24000, 47000, 36000, 20000]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Set figure size for all plots
sns.set(rc={'figure.figsize':(6,6)})

# Box Plot: Body Style vs Price
plt.figure(figsize=(6, 6))
sns.boxplot(x=df["body-style"], y=df["price"], palette="Set3").set_title('Body Style vs Price Box Plot')
plt.show()

# Scatter Plot: Horsepower vs Price with Body Style Hue
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df, x="horsepower", y="price", hue="body-style", palette="Set3", s=100).set_title('Horsepower vs Price Scatter Plot')
plt.show()

# Bar Plot: Average Price per Make
plt.figure(figsize=(6, 6))
sns.barplot(x=df["make"], y=df["price"], palette="Set3", estimator=sum).set_title('Total Price per Make Bar Plot')
plt.show()

# Violin Plot: Horsepower Distribution by Body Style
plt.figure(figsize=(10, 6))
sns.violinplot(x=df["body-style"], y=df["horsepower"], palette="Set3").set_title('Horsepower Distribution by Body Style Violin Plot')
plt.show()

# Pair Plot: Pairwise Relationships in the Dataset
sns.pairplot(df, hue="body-style", palette="Set3")
plt.suptitle('Pair Plot of Car Features', y=1.02)
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "city": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
    "latitude": [40.7128, 34.0522, 41.8781, 29.7604, 33.4484],
    "longitude": [-74.0060, -118.2437, -87.6298, -95.3698, -112.0740],
    "car_sales": [100000, 90000, 75000, 60000, 50000]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))

# Load a base map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plotting
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the base map
world.plot(ax=ax, color='lightgrey')

# Plot the bubble map
gdf.plot(ax=ax, kind='scatter', x='longitude', y='latitude', s=df['car_sales'] / 100, color='blue', alpha=0.6, edgecolor='k', linewidth=0.5)

# Add labels
for x, y, label in zip(df["longitude"], df["latitude"], df["city"]):
    ax.text(x, y, label, fontsize=12, ha='right')

# Title and show plot
plt.title('Car Sales by City Bubble Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample dataset
data = {
    "car_model": ["Model A", "Model B", "Model C"],
    "sales": [75000, 55000, 40000],
    "target": [80000, 60000, 45000],
    "range1": [0, 0, 0],  # Poor range starts from 0 for all models
    "range2": [40000, 30000, 20000],  # Poor to average threshold
    "range3": [70000, 50000, 35000]  # Average to good threshold
}

# Create a bullet graph
fig, ax = plt.subplots(figsize=(10, 6))

# Define positions for the bullet graphs
positions = range(len(data["car_model"]))

# Plot ranges
for i, pos in enumerate(positions):
    ax.broken_barh([(data["range1"][i], data["range2"][i] - data["range1"][i])], (pos - 0.4, 0.8), facecolors='lightgray')
    ax.broken_barh([(data["range2"][i], data["range3"][i] - data["range2"][i])], (pos - 0.4, 0.8), facecolors='gray')
    ax.broken_barh([(data["range3"][i], data["target"][i] - data["range3"][i])], (pos - 0.4, 0.8), facecolors='darkgray')

# Plot actual sales
ax.barh(positions, data["sales"], height=0.5, color='skyblue')

# Plot targets
for i, pos in enumerate(positions):
    ax.plot(data["target"][i], pos, 'o', color='red', markersize=8)

# Set labels
ax.set_yticks(positions)
ax.set_yticklabels(data["car_model"])
ax.set_xlabel('Sales')
ax.set_title('Bullet Graph for Car Sales Performance')

plt.show()
```
